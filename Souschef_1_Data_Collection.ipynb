```python
#imports
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import time

from recipe_scrapers import scrape_me
```

## Data Collection: Recipe Links Collection

```python
url = "https://www.allrecipes.com/recipes/"
req = requests.get(url)
print(req.status_code)
```

```python
#create soup instance
html = req.text
soup = BeautifulSoup(html, 'lxml')

#find all meall type classes
mealtype = soup.find("ul", class_='browse-hubs__subcategories')
links = mealtype.find_all('li')

#isolate for meal type links
meal_types = []

for x in links:
    
    url = x.a.get('href')
    
    meal_types.append(url)
    
meal_types
```

```python
#soup instance for the meal_type links
url = "https://www.allrecipes.com/recipes/76/appetizers-and-snacks/"

result_app = requests.get(url)
html = result_app.text
soup = BeautifulSoup(html, 'lxml')

#isolate for 1 recipe 
soup.find('article', class_ = 'fixed-recipe-card').div.a.get('href')
```

So this gave us the first link from the recipe card that we need, great! let's go and get the rest now.

```python
recipes_app = [x.div.a.get('href') for x in soup.findAll("article", class_ = "fixed-recipe-card")]

print(str(len(recipes_app)) + " fetched products URLs")
print("One example:")
recipes_app
```

### Scrape Page 1 through Page 50 of Appetizers/Snacks 

```python
base_url = 'https://www.allrecipes.com/recipes/76/appetizers-and-snacks/'

recipes = []

i = 1

for i in range(1,51):

    url = base_url + '?page={}'.format(i)

    #get the soup
    result = requests.get(url)
    html = result.text
    soup = BeautifulSoup(html, 'lxml')

    #get recipes
    recipe = [x.div.a.get('href') for x in soup.findAll("article", class_ = "fixed-recipe-card")]

    recipes.append(recipe)
    
    i = i + 1

print(str(len(recipes)) + " recipes")
print("Some examples:")
recipes
```

This has given us 1450 recipes links (29 posts/per page where we scraped 50 pages in total) for Appetizers/Snacks. Now that we know the code works, let's implement this code for all of our meal types other than appetizers: breakfast/brunch, desserts, dinner, and drinks.

### Scraper for all  Meal Types (Apps/Breakfast/Dessert/Dinner/Drinks)

```python
allrecipe_links = []

for meal in meal_types:
    
    base_url = meal

    i = 1

    for i in range(1,51):

        url = base_url + '?page={}'.format(i)

        #get the soup
        result = requests.get(url)
        html = result.text
        soup = BeautifulSoup(html, 'lxml')

        #get recipes
        recipe = [x.div.a.get('href') for x in soup.findAll("article", class_ = "fixed-recipe-card")]

        allrecipe_links.append(recipe)

        i = i + 1

print(str(len(allrecipe_links)) + " recipes for each Meal Type")
print("Let's see how it looks:")
allrecipe_links
```

```python
allrecipe_links
```

This has given us 5039 recipe links in total. We're now read to ingest these recipe links into our recipe scraper to return the specific recipe title, instructions, ingredients etc. for each respectice recipe.

## Data Collection: Specific Recipe Information

Using the recipe links that we scraped from _AllRecipes_ we can now use the recipe scraper created by [marcolussetti](https://github.com/marcolussetti/recipe-scraper) to scrape the relevant information (ie. Recipe name/instructions/ingredients etc.) from each of these recipe links.

```python
# give the url as a string, it can be url from any site listed below
scraper = scrape_me('https://www.allrecipes.com/recipe/19673/seven-layer-taco-dip/')

#scraper.title()
#scraper.total_time()
#scraper.ingredients()
scraper.instructions()
```

```python
#put all the recipes into a single list as it is currently list of lists
links = []

for i in allrecipe_links:
    for x in i:
        links.append(x)
```

```python
len(links)
```

```python
#run the links through the recipe scraper
recipes = []

for url in links:
    result = {}
    
    result['title'] = scrape_me(url).title()
    result['time'] = scrape_me(url).total_time()
    result['ingredients'] = scrape_me(url).ingredients()
    result['instructions'] = scrape_me(url).instructions()
    #result['link'] = url
    
    recipes.append(result)
    
    time.sleep(2)
    
```

```python
recipes
```

Create a pandas DataFrame from the list of recipes scraped:

```python
df = pd.DataFrame(recipes)
df.head()
```

```python
df.to_csv('recipes.csv', index = False)
```
